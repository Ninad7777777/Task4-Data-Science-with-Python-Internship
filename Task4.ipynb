```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Load the dataset
file_path = 'C:\\Users\\AKASH SINGH YADAV\\Downloads\\USvideos.csv'
data = pd.read_csv(file_path)
```

```python
# Convert infinite values to NaN
data.replace([float('inf'), float('-inf')], float('nan'), inplace=True)
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Suppress warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=FutureWarning)
    
    # Plot distributions of views, likes, dislikes, and comment counts
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # Views
    sns.histplot(data['views'], bins=50, kde=True, ax=axes[0, 0])
    axes[0, 0].set_title('Distribution of Views')

    # Likes
    sns.histplot(data['likes'], bins=50, kde=True, ax=axes[0, 1])
    axes[0, 1].set_title('Distribution of Likes')

    # Dislikes
    sns.histplot(data['dislikes'], bins=50, kde=True, ax=axes[1, 0])
    axes[1, 0].set_title('Distribution of Dislikes')

    # Comment Count
    sns.histplot(data['comment_count'], bins=50, kde=True, ax=axes[1, 1])
    axes[1, 1].set_title('Distribution of Comment Count')

    plt.tight_layout()
    plt.show()
```

```python
# Plot box plots of views, likes, dislikes, and comment counts to identify outliers
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Views
sns.boxplot(x=data['views'], ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Views')

# Likes
sns.boxplot(x=data['likes'], ax=axes[0, 1])
axes[0, 1].set_title('Box Plot of Likes')

# Dislikes
sns.boxplot(x=data['dislikes'], ax=axes[1, 0])
axes[1, 0].set_title('Box Plot of Dislikes')

# Comment Count
sns.boxplot(x=data['comment_count'], ax=axes[1, 1])
axes[1, 1].set_title('Box Plot of Comment Count')

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = data[['views', 'likes', 'dislikes', 'comment_count']].corr()
```

```python
# Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
